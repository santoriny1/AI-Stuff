```python
import cv2 as cv
import numpy as np
import pandas as pd
```

```python
img = cv.imread('image2.jpg', 1)
img2 = cv.cvtColor(img, cv.COLOR_BGR2HSV)

#Filtro *AMARILLO*
ubb = (25, 60, 60)
uba = (35, 255, 255)

mask= cv.inRange(img2, ubb, uba)

#Filtro entre 2 rangos *ROJO*
ubb1 = (0, 60, 60)
uba1 = (10, 255, 255)

ubb2 = (170, 60, 60)
uba2 = (180, 255, 255)

mask2= cv.inRange(img2, ubb1, uba1)
mask3= cv.inRange(img2, ubb2, uba2)

mask1= mask2 + mask3

#Filtro *NARANJA*
ubb3 = (11, 130, 130)
uba3 = (24, 255, 255)

mask4= cv.inRange(img2, ubb3, uba3)

#Filtro *VERDE*
ubb4 = (36, 60, 60)
uba4 = (84, 255, 255)

mask5= cv.inRange(img2, ubb4, uba4)

#Filtro *TURQUESA*
ubb5 = (85, 60, 60)
uba5 = (96, 255, 255)

mask6= cv.inRange(img2, ubb5, uba5)

#Filtro *AZUL*
ubb6 = (97, 60, 60)
uba6 = (124, 255, 255)

mask7= cv.inRange(img2, ubb6, uba6)

#Filtro *MORADO*
ubb7 = (125, 60, 60)
uba7 = (132, 255, 255)

mask8= cv.inRange(img2, ubb7, uba7)

#Filtro *ROSA*
ubb8 = (133, 60, 60)
uba8 = (169, 255, 255)

mask9= cv.inRange(img2, ubb8, uba8)

#Filtro *BLANCO*
ubb9 = (0, 0, 168)
uba9 = (172, 60, 255)

mask10= cv.inRange(img2, ubb9, uba9)

#Aplicando filtros
amarillo= cv.bitwise_and(img, img, mask=mask)
rojo= cv.bitwise_and(img, img, mask=mask1)
naranja= cv.bitwise_and(img, img, mask=mask4)
verde= cv.bitwise_and(img, img, mask=mask5)
turquesa= cv.bitwise_and(img, img, mask=mask6)
azul= cv.bitwise_and(img, img, mask=mask7)
morado= cv.bitwise_and(img, img, mask=mask8)
rosa= cv.bitwise_and(img, img, mask=mask9)
blanco= cv.bitwise_and(img, img, mask=mask10)

cv.imshow("img", img)
cv.imshow("amarillo", amarillo)
cv.imshow("rojo", rojo)
cv.imshow("naranja", naranja)
cv.imshow("verde", verde)
cv.imshow("turquesa", turquesa)
cv.imshow("azul", azul)
cv.imshow("morado", morado)
cv.imshow("rosa", rosa)
cv.imshow("blanco", blanco)

df = pd.DataFrame(np.transpose(mask.nonzero()), columns=["x","y"])
df1 = pd.DataFrame(np.transpose(mask1.nonzero()), columns=["x","y"])
df2 = pd.DataFrame(np.transpose(mask4.nonzero()), columns=["x","y"])
df3 = pd.DataFrame(np.transpose(mask5.nonzero()), columns=["x","y"])
df4 = pd.DataFrame(np.transpose(mask6.nonzero()), columns=["x","y"])
df5 = pd.DataFrame(np.transpose(mask7.nonzero()), columns=["x","y"])
df6 = pd.DataFrame(np.transpose(mask8.nonzero()), columns=["x","y"])
df7 = pd.DataFrame(np.transpose(mask9.nonzero()), columns=["x","y"])
df8 = pd.DataFrame(np.transpose(mask10.nonzero()), columns=["x","y"])

cv.waitKey(0)
cv.destroyAllWindows()
```
